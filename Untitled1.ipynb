```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# تحديد مسارات البيانات داخل Google Drive
train_dir = '/content/drive/MyDrive/dl/chest_xray/train'
test_dir = '/content/drive/MyDrive/dl/chest_xray/test'
val_dir = '/content/drive/MyDrive/dl/chest_xray/val'
```

```python
import os # Importing the os module
import pandas as pd

# Generate data paths with labels
train_data_dir = '/content/drive/MyDrive/dl/chest_xray/train'
test_dir = '/content/drive/MyDrive/dl/chest_xray/test'

# Initialize empty lists to store filepaths and labels
filepaths = []
labels = []

folds = os.listdir(train_data_dir)
# print(folds)

for fold in folds:
    foldpath = os.path.join(train_data_dir, fold)
    filelist = os.listdir(foldpath)
    for file in filelist:
        fpath = os.path.join(foldpath, file)

        filepaths.append(fpath)
        labels.append(fold)

# print(labels)

# Concatenate data paths with labels into one dataframe
Fseries = pd.Series(filepaths, name= 'filepaths')
Lseries = pd.Series(labels, name='labels')

train_df = pd.concat([Fseries, Lseries], axis= 1)
```

```python
# valid and test dataframe
from sklearn.model_selection import train_test_split # Import train_test_split
# Use train_df instead of ts_df
valid_df, test_df = train_test_split(train_df,  train_size= 0.50, shuffle= True, random_state= 123)
test_df
```

```python
# إعداد التحويلات بدون أي تغييرات
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# تحميل البيانات الأصلية
train_generator_before = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

test_generator_before = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)
```

```python
# إعداد التحويلات مع التعديلات (مثل التكبير والتقليب)
train_datagen_after = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# تحميل البيانات بعد التحويلات
train_generator_after = train_datagen_after.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

test_generator_after = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.regularizers import l2 # Import l2 from the correct module
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

def build_cnn_model():
    # Define img_shape with the desired input image shape
    img_shape = (150, 150, 3)  # Assuming input images are 150x150 pixels with 3 color channels

    # Now you can use img_shape in your model definition
    base_modle=tf.keras.applications.efficientnet.EfficientNetB3(include_top=False , weights='imagenet' , input_shape=img_shape , pooling='max')
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
        MaxPooling2D(pool_size=(2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# تدريب نموذج CNN مع البيانات قبل وبعد المعالجة
cnn_model = build_cnn_model()

# تدريب باستخدام البيانات الأصلية (قبل التحويلات)
history_cnn_before = cnn_model.fit(
    train_generator_before,
    steps_per_epoch=100,
    epochs=30,
    validation_data=test_generator_before,
    validation_steps=50,
    verbose=1
)

# تدريب باستخدام البيانات بعد التحويلات
history_cnn_after = cnn_model.fit(
    train_generator_after,
    steps_per_epoch=100,
    epochs=30,
    validation_data=test_generator_after,
    validation_steps=50,
    verbose=1
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam # Import Adam

def build_vgg16_model(input_shape=(150, 150, 3)):
    # Load VGG16 model pre-trained on ImageNet, without the top layers (fully connected layers)
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)

    # Build a new model by stacking custom layers on top of VGG16
    model = Sequential([
        base_model,  # Add the base VGG16 model
        GlobalAveragePooling2D(),  # Pool the output into a 1D vector
        Dense(512, activation='relu'),  # Add a fully connected layer
        Dropout(0.5),  # Dropout to reduce overfitting
        Dense(1, activation='sigmoid')  # Final output layer (binary classification)
    ])

    # Freeze the layers of VGG16 so they won't be trained
    for layer in base_model.layers:
        layer.trainable = False

    # Compile the model with the Adam optimizer and binary crossentropy loss
    model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy']) # Use Adam()

    return model
```

```python
# تدريب نموذج VGG16 مع البيانات قبل وبعد المعالجة
vgg16_model = build_vgg16_model()

# تدريب باستخدام البيانات الأصلية (قبل التحويلات)
history_vgg16_before = vgg16_model.fit(
    train_generator_before,
    steps_per_epoch=100,
    epochs=20,
    validation_data=test_generator_before,
    validation_steps=50,
    verbose=1
)

# تدريب باستخدام البيانات بعد التحويلات
history_vgg16_after = vgg16_model.fit(
    train_generator_after,
    steps_per_epoch=100,
    epochs=20,
    validation_data=test_generator_after,
    validation_steps=50,
    verbose=1
)
```

```python
# Ensure the following cell defining the function has been run before
# Cell content of ipython-input-1-77cb00659b9d:

from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Sequential # Import Sequential here
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

def build_resnet_model(input_shape=(150, 150, 3)):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(1, activation='sigmoid')  # For binary classification
    ])

    for layer in base_model.layers:
        layer.trainable = False  # Freeze the layers of the base model

    model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# تدريب نموذج ResNet50 مع البيانات قبل وبعد المعالجة
resnet50_model = build_resnet_model()

# تدريب باستخدام البيانات الأصلية (قبل التحويلات)
history_resnet50_before = resnet50_model.fit(
    train_generator_before,
    steps_per_epoch=100,
    epochs=20,
    validation_data=test_generator_before,
    validation_steps=50,
    verbose=1
)

# تدريب باستخدام البيانات بعد التحويلات
history_resnet50_after = resnet50_model.fit(
    train_generator_after,
    steps_per_epoch=100,
    epochs=20,
    validation_data=test_generator_after,
    validation_steps=50,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

# Function to plot a scatter plot comparing training accuracy vs validation accuracy
def plot_scatter_plot(history_list, model_names, transformation_statuses):
    # Create a figure for the scatter plot
    plt.figure(figsize=(10, 6))

    # Loop through the history list for each model and transformation
    for i, history in enumerate(history_list):
        model_name = model_names[i]
        transformation_status = transformation_statuses[i]

        # Extract training and validation accuracy for the model
        if 'accuracy' in history.history and 'val_accuracy' in history.history:
            train_accuracy = history.history['accuracy'][-1]  # Get the final training accuracy
            val_accuracy = history.history['val_accuracy'][-1]  # Get the final validation accuracy

            # Plot a point on the scatter plot
            plt.scatter(train_accuracy, val_accuracy, label=f'{model_name} ({transformation_status})', s=100, edgecolors='black', alpha=0.7)

    # Customizing the plot
    plt.title('Training Accuracy vs Validation Accuracy')
    plt.xlabel('Training Accuracy')
    plt.ylabel('Validation Accuracy')
    plt.legend(loc='best')
    plt.grid(True)

    # Show the plot
    plt.tight_layout()
    plt.show()

# Example usage:
history_list = [
    history_cnn_before, history_cnn_after,
    history_vgg16_before, history_vgg16_after,
    history_resnet50_before, history_resnet50_after,
    history_inceptionv3_before, history_inceptionv3_after
]

model_names = ['CNN', 'CNN', 'VGG16', 'VGG16', 'ResNet50', 'ResNet50', 'InceptionV3', 'InceptionV3']
transformation_statuses = ['Before Transformation', 'After Transformation', 'Before Transformation', 'After Transformation',
                            'Before Transformation', 'After Transformation', 'Before Transformation', 'After Transformation']

plot_scatter_plot(history_list, model_names, transformation_statuses)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot a grouped bar chart comparing training accuracy vs validation accuracy
def plot_bar_chart(history_list, model_names, transformation_statuses):
    # Number of models to plot
    num_models = len(model_names)

    # Prepare lists for training accuracy and validation accuracy
    train_accuracies = []
    val_accuracies = []

    # Loop through each history and extract training and validation accuracy
    for history in history_list:
        if 'accuracy' in history.history and 'val_accuracy' in history.history:
            # Extract final accuracy values (after the last epoch)
            train_accuracies.append(history.history['accuracy'][-1])
            val_accuracies.append(history.history['val_accuracy'][-1])

    # Set the width of each bar
    bar_width = 0.35

    # Generate the positions for the bars on the x-axis
    r = np.arange(num_models)

    # Create the bar chart
    plt.bar(r - bar_width/2, train_accuracies, color='skyblue', edgecolor='black', width=bar_width, label='Training Accuracy')
    plt.bar(r + bar_width/2, val_accuracies, color='lightgreen', edgecolor='black', width=bar_width, label='Validation Accuracy')

    # Adding labels and title
    plt.xlabel('Models', fontweight='bold', fontsize=14)
    plt.ylabel('Accuracy', fontweight='bold', fontsize=14)
    plt.title('Training Accuracy vs Validation Accuracy', fontsize=16)

    # Customize the x-axis with the model names
    plt.xticks(r, model_names)

    # Add a legend
    plt.legend()

    # Show the plot
    plt.tight_layout()
    plt.show()

# Example usage:
history_list = [
    history_cnn_before, history_cnn_after,
    history_vgg16_before, history_vgg16_after,
    history_resnet50_before, history_resnet50_after,
    history_inceptionv3_before, history_inceptionv3_after
]

model_names = ['CNN', 'CNN', 'VGG16', 'VGG16', 'ResNet50', 'ResNet50', 'InceptionV3', 'InceptionV3']
transformation_statuses = ['Before Transformation', 'After Transformation', 'Before Transformation', 'After Transformation',
                            'Before Transformation', 'After Transformation', 'Before Transformation', 'After Transformation']

plot_bar_chart(history_list, model_names, transformation_statuses)
```

```python
import matplotlib.pyplot as plt

# دالة لعرض نتائج التدريب بعد كل epoch
def plot_history(history, model_name, transformation_status):
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} Accuracy ({transformation_status})')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

# عرض النتائج لكل نموذج
plot_history(history_cnn_before, 'CNN', 'Before Transformation')
plot_history(history_cnn_after, 'CNN', 'After Transformation')

plot_history(history_vgg16_before, 'VGG16', 'Before Transformation')
plot_history(history_vgg16_after, 'VGG16', 'After Transformation')

plot_history(history_resnet50_before, 'ResNet50', 'Before Transformation')
plot_history(history_resnet50_after, 'ResNet50', 'After Transformation')

plot_history(history_inceptionv3_before, 'InceptionV3', 'Before Transformation')
plot_history(history_inceptionv3_after, 'InceptionV3', 'After Transformation')
```

```python
import matplotlib.pyplot as plt

# Function to plot a pie chart for final accuracy (training vs. validation)
def plot_accuracy_pie_chart(history, model_name, transformation_status):
    # Extract final training and validation accuracy values
    final_train_accuracy = history.history['accuracy'][-1]
    final_val_accuracy = history.history['val_accuracy'][-1]

    # Data for the pie chart
    labels = ['Training Accuracy', 'Validation Accuracy']
    sizes = [final_train_accuracy * 100, final_val_accuracy * 100]  # Convert to percentage

    # Colors for the pie slices
    colors = ['#66b3ff', '#ffcc99']

    # Plotting the pie chart
    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})

    # Add title
    plt.title(f'{model_name} Accuracy Pie Chart ({transformation_status})', fontsize=16)

    # Show the plot
    plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
    plt.show()

# Plotting pie charts for each model (before and after transformation)
plot_accuracy_pie_chart(history_cnn_before, 'CNN', 'Before Transformation')
plot_accuracy_pie_chart(history_cnn_after, 'CNN', 'After Transformation')

plot_accuracy_pie_chart(history_vgg16_before, 'VGG16', 'Before Transformation')
plot_accuracy_pie_chart(history_vgg16_after, 'VGG16', 'After Transformation')

plot_accuracy_pie_chart(history_resnet50_before, 'ResNet50', 'Before Transformation')
plot_accuracy_pie_chart(history_resnet50_after, 'ResNet50', 'After Transformation')

plot_accuracy_pie_chart(history_inceptionv3_before, 'InceptionV3', 'Before Transformation')
plot_accuracy_pie_chart(history_inceptionv3_after, 'InceptionV3', 'After Transformation')
```
